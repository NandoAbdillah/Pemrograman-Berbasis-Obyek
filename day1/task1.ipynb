## Task 1  : Kalkulator Sederhana

```python
# Kalkulator sederhana

def penjumlahan (a, b):
    return a + b

def pengurangan (a, b):
    return a - b


def perkalian (a, b) :
    return a * b

def pembagian (a, b) :
    return a / b
```

```python
angka1 = int(input("Masukkan angka pertama : "))
```

```python
print(angka1)
```

```python
angka2 = int(input("Masukkan angka kedua : "))
```

```python
print(angka2)
```

```python
print("Hasil penjumlahan : " , penjumlahan(angka1, angka2))
print("Hasil pengurangan : ", pengurangan(angka1, angka1))
print("Hasil perkalian : ", perkalian(angka1, angka2))
print("Hasil pembagian : ", pembagian(angka1, angka2))
```
