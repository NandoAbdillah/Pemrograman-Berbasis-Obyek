```python
data = []

for i in range(10) :
    data.append(int(input(f"Masukkan data ke {i+1} : ")))
    
```

```python
for d in data :
    print(d)
```

```python
def rerata () :
    return sum(data) / len(data)

def terbesar() :
    return max(data)

def terkecil() :
    return min(data)

def ganjil_genap() :
    jum_ganjil = 0
    jum_genap = 0
    
    for d in data :
        if d % 2 != 0 :
            jum_ganjil += 1
        else :
            jum_genap += 1
            
    return jum_ganjil, jum_genap
```

```python
print("Rata-rata : ", rerata())
print("Angka terbesar : ", terbesar())
print("Angka terkecil : ", terkecil())

ganjil, genap = ganjil_genap()
print("Jumlah angka ganjil : ", ganjil)
print("Jumlah angka genap : ", genap)
```
