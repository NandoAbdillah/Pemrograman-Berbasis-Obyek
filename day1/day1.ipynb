```python
umur = True
teks = """
Nama saya 
Rusdi Ngawi
"""
print(umur)
print(teks)

# Tipe Data
print(type(umur))
print(type(teks))
```

```python
# Indexing 
nama = "Nando Abdillah"

print(nama[4])
```

```python
list_data = [1, 2.0, "PBOJ"]
print(type(list_data))
list_data[0] = "Teknik Informatika"
print(list_data[2])
print(list_data[-3])
```

```python
# list 
list_data = ["laptop", "monitor", "mouse", "mousepad", "keyboard", "webcam", "microphone"]
print(list_data[2:6:2])
```

```python
tuple_data = (1, 2.0, "PBOJ")
print(type(tuple_data))
# tuple_data[0] = 10  # This line would raise an error because tuples are immutable
print(tuple_data[2])
print(tuple_data[-3])
```

```python
# Tipe Data Set

set_data = {1,2,3,4,5,6,7}

print(type(set_data))
print(set_data[0]) # not subscriptable
```

```python
# Tipe Data Dictionary
dict_data = {
    "nama" : "Nando Abdillah Salam",
    "umur" : 20,
    "jurusan" : "Teknik Informatika",
    "kawin" : False
}

print(type(dict_data))
# print(dict_data[0]) # not subscriptable
print(dict_data["nama"])
print(type(dict_data["kawin"]))
```
